### 1. IMPORT LIBRARIES AND LOAD DATASETS 


```python
 #1.Import Libraries:
  
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
  
```

```python
  # Example of loading a dataset
df = pd.read_csv('C:\\Users\\D.SURESH KUMAR\\Desktop\\archive1\\zomato.csv')
  
```

### 2. Clean and preprocessing data

```python

#Handle Missing Values:
df.dropna(inplace=True)  # Remove rows with missing values
  # or
df.fillna(method='ffill', inplace=True)  # Forward fill missing values
# Remove Duplicates: 
df.drop_duplicates(inplace=True)
  
```

```python
print(df.columns)
#Convert Data Types:
# Extract city from the 'listed_in(city)' column
df['city'] = df['listed_in(city)']
```

### 3. Conduct Exploratory Data Analysis (EDA)

```python

#Descriptive Statistics:
print(df.describe())
  
```

```python
#Visualizations:
#1.Distribution of Ratings:
plt.figure(figsize=(10, 6))
sns.histplot(df['rate'].dropna(), bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

#2. Online Order vs Book Table

plt.figure(figsize=(10, 6))
sns.countplot(x='online_order', hue='book_table', data=df)
plt.title('Online Order vs Book Table')
plt.xlabel('Online Order')
plt.ylabel('Count')
plt.show()

#3. Restaurant Types in Different Cities

plt.figure(figsize=(12, 8))
sns.countplot(y='rest_type', hue='listed_in(city)', data=df, order=df['rest_type'].value_counts().index)
plt.title('Restaurant Types in Different Cities')
plt.xlabel('Count')
plt.ylabel('Restaurant Type')
plt.legend(title='City')
plt.show()

#4. Top Cuisines

plt.figure(figsize=(12, 8))
top_cuisines = df['cuisines'].value_counts().head(10)
sns.barplot(x=top_cuisines, y=top_cuisines.index)
plt.title('Top 10 Cuisines')
plt.xlabel('Count')
plt.ylabel('Cuisine')
plt.show()

#5. Approximate Cost Distribution

plt.figure(figsize=(10, 6))
sns.histplot(df['approx_cost(for two people)'].dropna(), bins=20, kde=True)
plt.title('Approximate Cost Distribution')
plt.xlabel('Approximate Cost (for two people)')
plt.ylabel('Frequency')
plt.show()
```

### 4: Draw Conclusions and Insights

```python
 #Answer Business Questions:
 
  # Example: Finding top-selling items
#1. Top-Selling Items

top_items = df['menu_item'].value_counts().head(10)
print(top_items)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_items.values, y=top_items.index, palette='viridis')
plt.title('Top 10 Selling Items')
plt.xlabel('Count')
plt.ylabel('Menu Item')
plt.show()

#2. Most Liked Dishes

top_dishes = df['dish_liked'].value_counts().head(10)
print(top_dishes)

plt.figure(figsize=(12, 8))
sns.barplot(x=top_dishes.values, y=top_dishes.index, palette='viridis')
plt.title('Top 10 Liked Dishes')
plt.xlabel('Count')
plt.ylabel('Dish Liked')
plt.show()
### Additional Insights

#1. Selecting Numeric Columns:
  
numeric_df = df.select_dtypes(include=[np.number])
   
   
#2. Computing the Correlation Matrix:
  
corr_matrix = numeric_df.corr()

#3. Plotting the Correlation Matrix:
  
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
